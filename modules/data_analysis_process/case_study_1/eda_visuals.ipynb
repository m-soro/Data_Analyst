# EDA with Visuals
Create visualizations to answer the quiz questions below this notebook. Use the `winequality_edited.csv` dataset, which you should've created  in the previous concept: *Appending Data (cont.)*.

```python
# Load dataset
import pandas as pd

% matplotlib inline

df = pd.read_csv('winequality_edited.csv')
```

```python
df.columns
```

### Histograms for Various Features

```python
# Fixed Acidity
df['fixed_acidity'].plot(kind='hist',figsize=(6,6));
```

```python
# Total Sulfur Dioxide
df['total_sulfur_dioxide'].plot(kind='hist',figsize=(6,6));
```

```python
# pH
df['pH'].plot(kind='hist',figsize=(6,6));
```

```python
# Alcohol
df['alcohol'].plot(kind='hist',figsize=(6,6));
```

### Scatterplots of Quality Against Various Features

```python
# fixed acidity
df.plot(kind='scatter',x='quality',y='fixed_acidity',figsize=(6,6));
```

```python
# total_sulfur_dioxide
df.plot(kind='scatter',x='quality',y='total_sulfur_dioxide',figsize=(6,6));
```

```python
# pH
df.plot(kind='scatter',x='quality',y='pH',figsize=(6,6));
```

```python
# alcohol
df.plot(kind='scatter',x='quality',y='alcohol',figsize=(6,6));
```
