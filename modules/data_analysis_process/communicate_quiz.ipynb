```python
# imports and load data
import pandas as pd

% matplotlib inline

df = pd.read_csv('store_data.csv', parse_dates=['week'])
df.info()
```

```python
# explore data
df.hist(figsize=(8,8));
```

```python
# sales for the last month
summary = df.groupby([df['week'].dt.year, df['week'].dt.month]).sum()

ax = summary[-1:].plot(kind='bar',rot=0, colormap='Set2');
ax.set(xlabel="Year and Month", ylabel="Sales [$]",title='Last Month\'s Sales');
```

```python
# average sales
ax = df.mean().plot(kind='bar',rot=0, colormap='Set3');
ax.set(xlabel="Stores", ylabel="Sales [$]",title='Average Sales');
```

```python
# sales for the week of March 13th, 2016
specific_week = df[df['week']== 'March 13, 2016']

ax = specific_week.iloc[:,1:].plot(kind='bar', rot=0, colormap='Set3', 
                      title='Sales for the week of March 13th, 2016');

# hide just the x axis
x_axis = ax.axes.get_xaxis()
x_axis.set_visible(False)
```

```python
# share of sales for the latest 3-month periods
ax = summary[-3:].plot(kind='bar', colormap='Set3',rot=0)
ax.set(xlabel="Year and Month", ylabel="Sales [$]",title='Share ' 
                       'of sales for the latest 3-month periods');
```
