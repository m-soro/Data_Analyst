# Exploring with Visuals

Use histograms and scatterplots to explore `clean_08.csv` and `clean_18.csv` in the Jupyter notebook. Then, answer the quiz questions below the notebook.

```python
# load datasets
import pandas as pd

% matplotlib inline

df_08 = pd.read_csv('clean_08.csv')
df_18 = pd.read_csv('clean_18.csv')
```

```python
# A quick scatter matrix to view relationships
pd.plotting.scatter_matrix(df_08,figsize=(10,10));
```

### Question 1
Compare the distributions of greenhouse gas score in 2008 to 2018.

**->** Distribution for **2008** is more **skewed to the left**. 

```python
df_08['greenhouse_gas_score'].hist();
```

```python
df_18['greenhouse_gas_score'].hist();
```

### Question 2
How has the distribution of combined mpg changed from 2008 to 2018?

**->** Became much more **skewed to the right**. 

```python
df_08['cmb_mpg'].hist(alpha=0.5)
df_18['cmb_mpg'].hist(alpha=0.8);
```

### Question 3
Describe the correlation between displacement and combined mpg.

**->** **Negative correlation** - The higher the cmb_mpg, the lower the displ. 

```python
df_18.plot(kind='scatter', x='displ',y='cmb_mpg');
```

### Question 4
Describe the correlation between greenhouse gas score and combined mpg.

**->** **Positive correlation** - The higher the cmb_mpg, the higher the greenhouse gas score. 

```python
df_18.plot(kind='scatter',x='greenhouse_gas_score',y='cmb_mpg');
```
